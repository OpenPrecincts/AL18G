```python
import pandas as pd
import geopandas as gpd
import os
from shutil import copyfile

from op_verification.reference_data import (
    geoid_to_county_name,
    state_to_fips,
    state_to_state_po,
    state_abbreviation_to_state_name,
    state_fip_to_county_to_geoid,
)
```

```python
parent_dir = 'data/shapefiles/county_level_precinct_election_results/'
shp_file_paths = [parent_dir + county_file for county_file in os.listdir(parent_dir) if not county_file[0] == '.']
shp_file_paths
```

```python
gdf = pd.concat([gpd.read_file(shp) for shp in shp_file_paths]).pipe(gpd.GeoDataFrame)
gdf.head(2)
```

```python
gdf = gdf.reset_index()
gdf_file_name = 'merged_shapefile'
gdf.to_file(gdf_file_name)
```
